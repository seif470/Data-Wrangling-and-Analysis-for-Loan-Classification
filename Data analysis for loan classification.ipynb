```python
#
seventy_fifth = df2["TotalIncome"].quantile(0.75)
twenty_fifth = df2["TotalIncome"].quantile(0.25)
TotalIncome_iqr = seventy_fifth - twenty_fifth
```

```python
TotalIncome_iqr
```

```python
# upper threshold
upper = seventy_fifth + (1.5 * TotalIncome_iqr)
```

```python
# lower threshold
lower = twenty_fifth - (1.5 * TotalIncome_iqr)
```
